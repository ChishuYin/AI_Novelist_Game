```python
from pathlib import Path
from datasets import Dataset
import pandas as pd
```

```python
def split_text_into_chunks(text, max_len=512):
    chunks, chunk = [], ""
    for line in text.split("\n"):
        if not line.strip():
            continue
        if len(chunk) + len(line) < max_len:
            chunk += " " + line.strip()
        else:
            chunks.append(chunk.strip())
            chunk = line.strip()
    if chunk:
        chunks.append(chunk.strip())
    return chunks
```

```python
# 加载所有 txt 并拆分为段落
raw_dir = Path("data/cleaned_data")
data = []

for file in raw_dir.glob("*.txt"):
    with file.open(encoding="utf-8") as f:
        chunks = split_text_into_chunks(f.read())
        for chunk in chunks:
            data.append({"text": chunk, "source": file.stem})
```

```python
# 转为 HuggingFace Dataset
df = pd.DataFrame(data)
dataset = Dataset.from_pandas(df) # 转为 HuggingFace Dataset
dataset.save_to_disk("data/processed/novel_dataset")
print("✅ 数据集构建完成，共包含段落数：", len(dataset))
```

```python
from datasets import load_from_disk

# 加载本地保存的数据集
dataset = load_from_disk("data/processed/novel_dataset")


```

```python
# 定义一个函数 clean_text，接收每一条样本 example（例如：{"text": "段落内容"}）
def clean_text(example):
    # 将 text 字段中的 '\xa0' 全部替换为空字符串 ''
    example["text"] = example["text"].replace('\xa0', '')
    return example

# 将这个清洗函数应用到整个 dataset（Dataset 类型，HuggingFace 格式）
dataset = dataset.map(clean_text)


```
